###### Sebuah kendaraan memiliki kecepatan awal 0 km/jam. Kendaraan tersebut dapat berjalan dengan menambah kecepatan, melambat dengan mengurangi kecepatan, dan berhenti apabila kecepatannya mencapai 0 km/jam. Program simulasi kendaraan ini harus berjalan terus-menerus hingga pengguna memilih untuk keluar dari program. Pengguna akan diberikan beberapa pilihan menu untuk mengontrol kendaraan, seperti menambah kecepatan, mengurangi kecepatan, melihat status kendaraan, atau keluar dari program. Buatlah sebuah program Python untuk mensimulasikan kondisi tersebut dengan menggunakan while loop agar program terus berjalan sampai pengguna memilih menu keluar. Tampilkan menu pilihan sebagai berikut: 1 untuk menambah kecepatan, 2 untuk mengurangi kecepatan, 3 untuk melihat status kendaraan, dan 4 untuk keluar dari program. Gunakan percabangan if-elif-else untuk memproses pilihan yang dimasukkan pengguna. Ketentuan dalam program adalah kecepatan maksimal kendaraan sebesar 120 km/jam dan kecepatan tidak boleh kurang dari 0 km/jam. Jika kecepatan sama dengan 0, tampilkan pesan "Kendaraan berhenti", sedangkan jika kecepatan lebih dari 0, tampilkan pesan "Kendaraan berjalan")).


```python
# Menyiapkan variabel awal
kecepatan = 0
kecepatan_maksimal = 120
print("Sistem kendaraan siap. Kecepatan awal:", kecepatan, "km/jam")
```

```python
def tampilkan_menu():
    print("\n=== MENU KENDARAAN ===")
    print("1. Tambah Kecepatan")
    print("2. Kurangi Kecepatan")
    print("3. Lihat Status Kendaraan")
    print("4. Keluar")

tampilkan_menu()
```

```python

while True:
    tampilkan_menu()
    pilihan = input("Masukkan pilihan Anda: ")

    if pilihan == '1':
        tambah = int(input("Tambah berapa km/jam? "))
        kecepatan += tambah
        if kecepatan > kecepatan_maksimal:
            kecepatan = kecepatan_maksimal
        print(f"Berhasil ditambah! Kecepatan: {kecepatan} km/jam")

    elif pilihan == '2':
        kurang = int(input("Kurangi berapa km/jam? "))
        kecepatan -= kurang
        if kecepatan < 0:
            kecepatan = 0
        print(f"Berhasil dikurangi! Kecepatan: {kecepatan} km/jam")

    elif pilihan == '3':
        print(f"\nSTATUS: {kecepatan} km/jam")
        if kecepatan == 0:
            print("Pesan: Kendaraan berhenti")
        else:
            print("Pesan: Kendaraan berjalan")

    elif pilihan == '4':
        print("Program dimatikan. Sampai jumpa!")
        break

    else:
        print("Pilihan salah, coba lagi.")
```
